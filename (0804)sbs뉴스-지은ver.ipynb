```python
import pandas as pd
df = pd.read_csv('sbs_eco_info.csv',encoding='utf-8')
sbs_eco_posts = df.get('0')

from konlpy.tag import Mecab
tagger = Mecab(dicpath="C:\\mecab\\mecab-ko-dic")

```

```python
sbs_eco_stop_words = "오늘 키워드 만 이슈"
sbs_eco_stop_words = sbs_eco_stop_words.split(' ')
sbs_eco_stop_words[0:10]

sbs_eco_nouns = []
for post in sbs_eco_posts:
    for noun in tagger.nouns(post):
        if noun not in sbs_eco_stop_words:
            sbs_eco_nouns.append(noun)

sbs_eco_nouns
```

```python
from collections import Counter
num_top_nouns = 20
sbs_eco_nouns_counter = Counter(sbs_eco_nouns)
sbs_eco_top_nouns = dict(sbs_eco_nouns_counter.most_common(num_top_nouns))
print(sbs_eco_top_nouns)

from wordcloud import WordCloud
sbs_eco_wc = WordCloud(background_color="white", font_path='./font/malgun.ttf')
sbs_eco_wc.generate_from_frequencies(sbs_eco_top_nouns)

import matplotlib.pyplot as plt
figure = plt.figure()
figure.set_size_inches(10, 20)
ax = figure.add_subplot(1, 1, 1)
ax.axis("off")
ax.imshow(sbs_eco_wc)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# 1글자도 인식이 되도록 토큰 패턴 변경
cv = CountVectorizer(token_pattern = r"(?u)\b\w+\b")
cv.fit(sentences)

def convert_cv(sentence, cv):
    
    # 문장을 토큰으로 분리
    tokenizer = Okt()
    tokens = tokenizer.morphs(sentence)
    
    # 토큰을 문자열로 변환
    sentence = " ".join(tokens)
    
    # CountVectorizer의 입력에 맞게 배열로 변경
    sentences = []
    sentences.append(sentence)
    
    # 벡터 변환
    vector = cv.transform(sentences).toarray()    
    
    return vector
```
