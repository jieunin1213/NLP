```python
import pandas as pd
df = pd.read_csv('sbs_eco_info.csv',encoding='utf-8')
sbs_eco_posts = df.get('0')

from konlpy.tag import Mecab
tagger = Mecab(dicpath="C:\\mecab\\mecab-ko-dic")

```

```python
sbs_eco_stop_words = "오늘 키워드 만 이슈"
sbs_eco_stop_words = sbs_eco_stop_words.split(' ')
sbs_eco_stop_words[0:10]

sbs_eco_nouns = []
for post in sbs_eco_posts:
    for noun in tagger.nouns(post):
        if noun not in sbs_eco_stop_words:
            sbs_eco_nouns.append(noun)

sbs_eco_nouns
```

```python
# 토큰을 문자열로 변환
sentence = " ".join(tokens)

# CountVectorizer의 입력에 맞게 배열로 변경
sentences = []
sentences.append(sentence)



sentences
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# 1글자도 인식이 되도록 토큰 패턴 변경
cv = CountVectorizer(token_pattern = r"(?u)\b\w+\b")
cv.fit(sentences)

def convert_cv(sentence, cv):
    
    # 문장을 토큰으로 분리
    tokenizer = Okt()
    tokens = tokenizer.morphs(sentence)
    
    # 토큰을 문자열로 변환
    sentence = " ".join(tokens)
    
    # CountVectorizer의 입력에 맞게 배열로 변경
    sentences = []
    sentences.append(sentence)
    
    # 벡터 변환
    vector = cv.transform(sentences).toarray()    
    
    return vector
```

```python
from numpy import dot
from numpy.linalg import norm
import numpy as np
from konlpy.tag import Okt
tokenizer = Okt()  
def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))
bigNum=0
temp_string=""
def comParedTo(kbsString):
    bigNum=0
    for post in sbs_eco_posts:
        temp_num=cos_sim(convert_cv(kbsString,cv)[0],convert_cv(post,cv)[0])
        if(temp_num>bigNum):
            bigNum=temp_num
            temp_string=post
    print(temp_string)
ass='중국'
comParedTo(ass)
```

```python
for line in lines : 
    # 단어의 기본형 사용
    malist = twitter.pos(line, norm=True, stem=True) # line을 pos(품사)로 태깅을 해준다.
    
    r = []
    for word in malist :
        # 어미/조사/구두점 등은 대상에서 제외 
        if not word[1] in ["Josa", "Eomi", "Punctuation"]:
            r.append(word[0]) # 단어들만 list에 담아주겠다!! (즉, ㅇ ㅓ미, 조사, 구두점 은 담지 않겠다! 제외하겠다1)
    rl = (" ".join(r)).strip()
    result.append(rl)
# print(result)


result
```

```python
word_file = 'sbs.word'
with open(word_file, 'w', encoding='utf-8') as fp :
    fp.write('\n'.join(result)) 
```

```python
import matplotlib
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc
from pandas.core.frame import DataFrame
```

```python
data = word2vec.LineSentence(word_file) # gensim에서 제공하는 LineSentence

model = word2vec.Word2Vec(data, size=200, 
                window=10, hs=1, min_count=2, sg=1)

model_name = 'sbs.model' 

model.save( model_name )

print('파일 ', model_name, '저장 완료')
```

```python
def showGraph( somedata ):
    font_location = 'c:/Windows/fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=font_location).get_name()
    matplotlib.rc('font', family=font_name) 
    
    su = len(somedata) # 전체 데이터 수
    
    # 축에 보여질 항목 이름들
    item = list( item[0] for item in somedata)
    
    # 그려지는 수치 데이터 
    count = list( item[1] for item in somedata)

    plt.barh(range(su), count, align='center')
    plt.yticks(range(su), item, rotation='10')
    plt.xlim(0.8, 0.9)
    plt.grid(True)  
    plt.show()    
```
