```python
from konlpy.tag import Komoran
komoran = Komoran()
```

```python
for aa in temp_list:
    if aa==' ' or aa=='  ':
        temp_list.remove(aa)
```

```python
haha=''
for num in temp_list:
    haha+=num
```

```python
print(komoran.morphs(haha))
```

```python
topWord=komoran.pos(haha)
```

```python
word_dic = {}
for word in topWord:
    if word[1] == "NNP": #  명사 확인하기 --- (※3)
        if not (word[0] in word_dic):
            word_dic[word[0]] = 0
        word_dic[word[0]] += 1 # 카운트하기


print(word_dic)
```

```python
# 값의 수치가 가장 큰 것 부터 역순으로 정렬하여 보여 준다.
keys = sorted(word_dic.items(), key=lambda x:x[1], reverse=True)
print(keys)
```

```python

df_a = pd.DataFrame.from_records(keys)
df_a.to_excel('(200803)김상일.xlsx')
```
